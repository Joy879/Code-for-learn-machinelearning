```
import pandas as pd
```

```
tData = pd.read_csv("/content/new_train.csv")
```

```
tData.head()
```

```
def getStock(df, number):
  '''
    Returns features and labels for each stock
  '''

  features = df[['Open-Stock-'+str(number), 'High-Stock-'+str(number), 'Low-Stock-'+str(number), 'VWAP-Stock-'+str(number)]]
  try:
    label = df['Close-Stock-'+str(number)]
    return features, label
  except:
    return features
```

```
stock1Feats, stock1Label = getStock(tData, 1)
stock2Feats, stock2Label = getStock(tData, 2)
stock3Feats, stock3Label = getStock(tData, 3)
stock4Feats, stock4Label = getStock(tData, 4)
stock5Feats, stock5Label = getStock(tData, 5)
```

```
training_set = stock1Feats.iloc[:280, :]
test_set = stock1Feats.iloc[280:, :]

training_set_y = stock1Label.iloc[:280]
test_set_y = stock1Label.iloc[280:]
```

```
df_train_norm = training_set.copy()
df_test_norm = test_set.copy()
```

```
from lazypredict.Supervised import LazyClassifier, LazyRegressor
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

##Regression

```
# fit all models
reg = LazyRegressor(predictions=True)
models, predictions = reg.fit(df_train_norm, df_test_norm, training_set_y, test_set_y)
```

##Classification

```
from sklearn.datasets import load_iris
```

```
data = load_iris()
X = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.2,random_state =0)
classifier = LazyClassifier(predictions=True)
models,predictions = classifier.fit(X_train, X_test, y_train, y_test)
```
